```python
ratings_with_name = ratings.merge(books,on='ISBN')
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df.columns = ['Book-Title', 'avg_rating']
print(avg_rating_df)
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('avg_rating',ascending=False).head(50)
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_rating']]
```

```python
popular_df.shape
```

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
plus_200 = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(plus_200)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index='Book-Title', columns='User-ID',values='Book-Rating')
```

```python
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
def recommend(book_name):
    index = np.where(pt.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])), key=lambda x: x[1], reverse=True)[1:6]
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        if not temp_df.empty:
            # Drop duplicates based on 'Book-Title'
            unique_books = temp_df.drop_duplicates(subset='Book-Title')
            item.append(unique_books['Book-Title'].values[0])
            item.append(unique_books['Book-Author'].values[0])
            item.append(unique_books['Image-URL-M'].values[0])
            data.append(item)
    return data
```

```python
recommend('1984')
```

```python
users_with_100_ratings = anime_ratings['user_id'].value_counts()
top_users = users_with_100_ratings[users_with_100_ratings > 100].index

top__users = top_users[:50000]

filtered_ratings = anime_ratings[anime_ratings['user_id'].isin(top__users)]

pivot_table = filtered_ratings.pivot_table(values='rating', 
                                           index='anime_id', 
                                           columns='user_id',
                                           fill_value=0)
```

```python
import pickle
pickle.dump(pivot_table,open('pivot_table_new.pkl','wb'))
```

```python
anime_SC = cosine_similarity(pivot_table)
```

```python
def recommend_anime(anime_name):
    anime_num = animes.loc[animes['Name'] == anime_name,'MAL_ID'].values[0]
    index = np.where(pivot_table.index == anime_num)[0][0]
    similar_items = sorted(list(enumerate(anime_SC[index])), key=lambda x: x[1], reverse=True)[1:6]
    data = []
    for i in similar_items:
        item = []
        temp_df = animes[animes['MAL_ID'] == pivot_table.index[i[0]]]
        if not temp_df.empty:
            # Drop duplicates based on 'Book-Title'
            unique_anime = temp_df.drop_duplicates(subset='MAL_ID')
            item.append(unique_anime['Name'].values[0])
            item.append(unique_anime['Score'].values[0])
            item.append(unique_anime['images'].values[0])
            item.append(unique_anime['Episodes'].values[0])
            data.append(item)
    return data
```
